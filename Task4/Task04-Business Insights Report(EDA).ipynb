Conversion Distribution

```python
df['Conversion'].value_counts()
```

```python
sns.countplot(x='Conversion', data=df)
plt.title('Conversion Distribution')
plt.show()
```

Gender Analysis

```python
sns.countplot(x='Gender', hue='Conversion', data=df)
plt.title('Conversion by Gender')
plt.show()
```

Campaign Channel Performance

```python
channel = df.groupby('CampaignChannel')['Conversion'].mean().sort_values(ascending=False)

channel.plot(kind='bar')
plt.title('Conversion Rate by Campaign Channel')
plt.ylabel('Average Conversion')
plt.show()
```

Campaign Type Analysis

```python
campaign = df.groupby('CampaignType')['Conversion'].mean().sort_values(ascending=False)

campaign.plot(kind='bar')
plt.title('Conversion Rate by Campaign Type')
plt.ylabel('Average Conversion')
plt.show()
```

Ad Spend Analysis

```python
sns.boxplot(x='Conversion', y='AdSpend', data=df)
plt.title('Ad Spend vs Conversion')
plt.show()
```

Website Visits Analysis

```python
sns.boxplot(x='Conversion', y='WebsiteVisits', data=df)
plt.title('Website Visits vs Conversion')
plt.show()
```

Email Opens Analysis

```python
sns.boxplot(x='Conversion', y='EmailOpens', data=df)
plt.title('Email Opens vs Conversion')
plt.show()
```

Email Clicks Analysis

```python
sns.boxplot(x='Conversion', y='EmailClicks', data=df)
plt.title('Email Clicks vs Conversion')
plt.show()
```

Income Distribution

```python
plt.figure(figsize=(8,5))

sns.histplot(df['Income'], bins=30, kde=True)

plt.title('Income Distribution')
plt.show()
```

Correlation Heatmap

```python
plt.figure(figsize=(12,8))

sns.heatmap(
    df.select_dtypes(include=np.number).corr(),
    annot=True,
    cmap='coolwarm'
)

plt.title('Correlation Heatmap')
plt.show()
```

Top Factors Affecting Conversion

```python
corr = df.select_dtypes(include=np.number).corr()['Conversion'].sort_values(ascending=False)

print(corr)
```

Funnel Analysis

```python
visits = df['WebsiteVisits'].sum()
opens = df['EmailOpens'].sum()
clicks = df['EmailClicks'].sum()
conversions = df['Conversion'].sum()

funnel = pd.DataFrame({
    'Stage':['Website Visits','Email Opens','Email Clicks','Conversions'],
    'Count':[visits,opens,clicks,conversions]
})

funnel
```

```python
plt.figure(figsize=(8,5))

plt.bar(funnel['Stage'], funnel['Count'])

plt.title('Marketing Funnel')

plt.show()
```

Channel Wise Conversion %

```python
channel_conversion = (
    df.groupby('CampaignChannel')['Conversion']
    .mean()
    .sort_values(ascending=False)*100
)

print(channel_conversion)
```
